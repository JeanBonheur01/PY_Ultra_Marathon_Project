```python
df[df['Event name'] == 'Everglades 50 Mile Ultra Run (USA)']['Event name'].str.split('(').str.get(1).str.split(')').str.get(0)
```

##### Combine all the filters togheter 

```python
df[(df['Event distance/length'].isin(['50km' , '50mi'])) & (df['Year of event'] == 2020) & (df['Event name'].str.split('(').str.get(1).str.split(')').str.get(0) == 'USA')]
```

```python
df2 = df[(df['Event distance/length'].isin(['50km' , '50mi'])) & (df['Year of event'] == 2020) & (df['Event name'].str.split('(').str.get(1).str.split(')').str.get(0) == 'USA')]
```

```python
df2.shape
```

##### Since all data is from USA race, remove (USA) from event name

```python
df2['Event name'].str.split('(').str.get(0)
```

```python
#df2['Event name'] = df2['Event name'].str.split('(').str.get(0)  or
df2.loc[:, 'Event name'] = df2['Event name'].str.split('(').str.get(0)
```

##### .loc[row_indexer,col_indexer] = value instead    - This is used when trying to replace correctly a value in the DataFrame


```python
df2.head(10)
```

##### clean up athlet age (removing the unwanted 0 value at the end)

```python
df2.loc[:, 'Athlete year of birth'] = df2['Athlete year of birth'].fillna(0)
```

```python
df2.loc[:, 'athlete_age'] = 2020 - df2['Athlete year of birth'].astype(int)
```

##### Remove (h) from athlete performance

```python
df2.loc[:, 'Athlete performance'] = df2['Athlete performance'].str.split(' ').str.get(0)
```

```python
df3 = df2.drop(['Athlete club', 'Athlete country', 'Athlete year of birth', 'Athlete age category'], axis =1)
```

##### 2. Which age group runs faster and which runs slower in the 50-mile races?

```python
df4.query('race_distance == "50mi"').groupby('age') ['average_speed'].agg (['mean', 'count']).sort_values('mean', ascending = False).query('count>19')
#shows the fastest runners, count 19 include only group of age were we have more than 19 records
```

```python
df4.query('race_distance == "50mi"').groupby('age') ['average_speed'].agg (['mean', 'count']).sort_values('mean', ascending = True).query('count>19')
#shows the lowest runners
```

##### Finding 7: Athletes who are 29 years old are the fastest, with the highest mean in the dataset, followed by those who are 23, 28, 30, and 25 years old. Overall, athletes between 25 and 30 years old are the fastest in the race. Athletes between of 50 and above have the lowest average speed.

###### Split between two decimals in the race_date

```python
df4['race_month']= df4['race_date'].str.split('.').str.get(1).astype(int)
```

```python
df4['race_season'] = df4['race_month'].apply(lambda x: 'Winter' if x > 11 else 'Fall' if x > 8 else 'Summer' if x > 5 else 'Sprint' if x >2 else 'Winter')
df4.head()
```

##### The month was extracted from the race_date, and then conditions was used in the query to determine the seanson of the race. 

```python
df4.groupby('race_season')['average_speed'].agg(['mean', 'count']).sort_values('mean', ascending = False)
```

##### Finding 8: Athletes perform better and faster in the spring, followed by winter and fall. However, their performance is lower in the summer when it is warmer."

##### What is the performance rate in the 50-mile races? 

```python
df4.query('race_distance == "50mi"').groupby('race_season')['average_speed'].agg(['mean', 'count']).sort_values('mean', ascending = False)
```

##### Finding 9: For 50-mile races, summer still remains the season where athletes perform the lowest, but the ranking has changed for other seasons. Fall has the highest performance rate, followed by spring and winter.
